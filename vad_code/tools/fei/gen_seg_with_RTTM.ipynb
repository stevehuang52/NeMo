## combine

```python

X = []
listen_jsons = glob.glob("/home/fjia/code/manifests_ami/ami_manifest_test/*.json")

speech = 0
background = 0
for one in listen_jsons:
    for line in open(one, 'r'):
        X.append(json.loads(line))
        if json.loads(line)['label'] == "speech":
            speech += 1
        else:
            background += 1
            
print(len(X), speech, background, speech+background)
```

```python
with open("/home/fjia/code/manifests_ami/new_ami_manifest_test.json", 'a') as fout:
    for i in X:
        json.dump(i, fout)
        fout.write('\n')
        
```
